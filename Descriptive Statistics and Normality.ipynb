```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
# functions for repetitive activities
def create_distribution(dist_list, desired_dist, datapoints):
    """creates numpy distributions for assignment"""
    if desired_dist == dist_list[0]:
        repeated = int(input('How many times is the action repeated (as an int)? '))
        probability = float(input('What is the probability of success per action (as a float)? '))
        binomial = np.random.binomial(repeated, probability, datapoints)
        return binomial
    elif desired_dist == dist_list[1]:
        scale = float(input('Enter a number for scale (as an int or float)? '))
        exponential = np.random.exponential(scale, datapoints)
        return exponential
    elif desired_dist == dist_list[2]:
        scale = float(input('Enter a starting mean (as an int or float)? '))
        decay = float(input('Enter a figure representing decay (as an int or float)? '))
        laplace = np.random.laplace(scale, decay, datapoints)
        return laplace
    elif desired_dist == dist_list[3]:
        shape = float(input('Enter a number greater than 0 (as an int or float)? '))
        pareto = np.random.pareto(shape, datapoints)
        return pareto
    elif desired_dist == dist_list[4]:
        lam = float(input('Enter an interval >=0 (as an int or float)? '))
        poisson = np.random.poisson(lam, datapoints)
        return poisson
    elif desired_dist == dist_list[5]:
        lower = float(input('Enter a lower bound (as an int or float)? '))
        upper = float(input('Enter an upper bound (as an int or float)? '))
        uniform = np.random.uniform(lower, upper, datapoints)
        return uniform
    else:
        print('Error.  Please check your code and/or inputs.')


def create_histograms(distribution, title):
    """creates histogram to visualize distributions"""
    # set chart sizes
    plt.figure(figsize=(10, 5))
    
    # plot 'normal' histogram of the data
    plt.subplot(1, 2, 1)
    plt.hist(distribution, color='b')
    plt.title(title)
    
    # plot the histogram with the mean and standard deviation included
    plt.subplot(1, 2, 2)
    plt.hist(distribution, color='c')
    plt.title(title + ' Mean and Standard Deviation')
    
    # calculate mean and standard deviation
    the_mean = distribution.mean()
    stddev = distribution.std()
    
    # add vertical lines for mean and standard deviation
    plt.axvline(the_mean, color='b', linestyle='solid', linewidth=2)
    plt.axvline(distribution.mean() + stddev, color='b', linestyle='dashed', linewidth=2)
    plt.axvline(distribution.mean() - stddev, color='b', linestyle='dashed', linewidth=2)    
    # plot graphs
    plt.show()
```

```python
# The selected distributions
distributions_list = [
    'Binomial',
    'Exponential',
    'Laplace',
    'Pareto',
    'Poisson',
    'Uniform',
    'Normal',
]
```

### Binomial Analysis

```python
graph_title = 'Binomial Distribution'
binomial = create_distribution(distributions_list, distributions_list[0], 100)
create_histograms(binomial, graph_title)
```

```python
graph_title = 'Exponential Distribution'
exponential = create_distribution(distributions_list, distributions_list[1], 100)
create_histograms(exponential, graph_title)
```

```python
graph_title = 'Laplace Distribution'
laplace = create_distribution(distributions_list, distributions_list[2], 100)
create_histograms(laplace, graph_title)
```

```python
graph_title = 'Pareto Distribution'
pareto = create_distribution(distributions_list, distributions_list[3], 100)
create_histograms(pareto, graph_title)
```

```python
graph_title = 'Poisson Distribution'
poisson = create_distribution(distributions_list, distributions_list[4], 100)
create_histograms(poisson, graph_title)
```

```python
graph_title = 'Uniform Distribution'
uniform = create_distribution(distributions_list, distributions_list[5], 100)
create_histograms(uniform, graph_title)
```

```python
# create 2 normally-distributed variables
norm1 = np.random.normal(5, 0.5, 1000)
norm2 = np.random.normal(10, 1, 1000)
norm1.sort()
norm2.sort()

# add them together to make a third variable
norm3 = norm1 + norm2

# graph histograms
graph_title = 'Normal'
create_histograms(norm3, graph_title)
```
